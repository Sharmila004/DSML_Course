# 0. an end-to-end scikit-learn workflow

### 1. get the data ready

```python
# stardard imports
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
heart_disease = pd.read_csv("./data/heart-disease.csv")
heart_disease 
```

```python
# create X (feature matrix)
x = heart_disease.drop ("target", axis = 1)

# create Y (labels)
y = heart_disease["target"]
```

### 2. choosing the right model and hyoerparameters

```python
# classification machine learning model
# capable of learning patterns and data and then classifying whenther a sample is one (a row) or another 
#(i.e., this or that)
from sklearn.ensemble import RandomForestClassifier 
# clf - classifier in scikit learn (ALT: model) 
# instantiate the classifier class in clf
clf = RandomForestClassifier()
# keep the default hyperparameters
clf.get_params()
```

### 3. fit the model to data

```python
from sklearn.model_selection import train_test_split

# split out data from x and y into x_train (training data), y_train, x_test (testing data), y_test
# test_size 0.2 => 20% ; that means 80% of data for training
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2)
```

```python
# ask classification model (randomForest) to find patterns in the training set of data
clf.fit(x_train, y_train)
```

```python
# 80% data which was set as training data and patterns were asked to be found in input data
x_train
```

```python
# 20% data which was set as testing data
x_test
```

```python
# make a prediction
import numpy as np
# y_label = clf.predict(np.array([0, 2, 3, 4]))
```

```python
# now that the patterns have been recognised in x_train ,
# using those patterns we are asking it to predict using the x_test data and save it as y predictions
y_preds = clf.predict(x_test)
y_preds
```

```python
# the y prediction (y_predict) seem to be simmilarto y_test with 0s and 1s
y_test
```

##### note: we can see that the predictions of y (y_preds) made by predicting the x_test data (similar to x_train) is similar to the expecrted y data (y_test)

### 4. evaluate the model

```python
# return mean accuracy of the given test data and labels
# shape of data must match
clf.score(x_train, y_train)
```

```python
clf.score(x_test, y_test)
```

```python
# functions for evaluating the performance of classification models.
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score
print(classification_report(y_test, y_preds)) # true labels vs predictions
```

```python
confusion_matrix(y_test, y_preds)
```

```python
accuracy_score(y_test, y_preds)
```

### 5. improve a medel

```python
# try different amount of n estimators (one of the parameters to try and immprove)
# eg:
np.random.seed(42)
for i in range (10, 100, 10):
    print(f"Trying model with {i} estimators") # f denotes f-string: formatted string literal
    clf = RandomForestClassifier(n_estimators = i).fit(x_train, y_train) # adjusting parameters
    print(f"Model accurecy on test set: {clf.score(x_test, y_test) * 100:.2f}%") # shrtcut in py using f-string syntax: similar to round(num, 2)
    print(" ")
```

##### we can see that by adjusting the hyper parameters of our model from 10 (default) to 20, we are able to ncrese score from the actual accuracy to teh l;ast iterated accuracy


### 6. save the model and load it 

```python
# save model with py's pickle library
import pickle

pickle.dump(clf, open("./exported/random_forest_model_1.pkl", "wb")) # write in binary
```

```python
# load the saved model
loaded_model = pickle.load(open("./exported/random_forest_model_1.pkl", "rb")) # read binary
loaded_model.score(x_test, y_test)
```

##### note: the loaded model is observed to have improved upto the last iterated estimator... the actual accuracy to teh l;ast iterated accuracy
